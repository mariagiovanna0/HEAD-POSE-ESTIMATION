Script per analizzare l'intero insieme dei volti estratti dal dataset, al fine di trovare le dimensioni massime e minime per effettuare un'analisi preliminare alla fase di resize bidimensionale.

```python
import os
import cv2

#path alla cartella contenente i volti estratti dal dataset
path_dataset="/content/drive/Shareddrives/Progetto/Colab Notebooks/volti estratti"

#funzione per trovare la dimensione minima
def find_min_size(path):
    min_dimension=16384
    min_couple=(128,128)

    #cerco per ogni immagine nelle sottocartelle della cartella generale la dimensione corrente e confronto con la minima
    for folder in os.listdir(path):
        #print("folder:", folder)
        for item in os.listdir(path+"/"+folder):
            #print(os.path.join(path+"/"+folder,item))
            image = cv2.imread(os.path.join(path+"/"+folder,item))

            annotated_image=image.copy()
            h,w,c=annotated_image.shape

            current_dimension=h*w
            current_couple=(h,w)

            #print("current_dimension:", current_dimension,"current_couple:", current_couple)
            if(current_dimension<min_dimension):
                min_dimension=current_dimension
                min_couple=current_couple

    #restituisco la dimensione totale, il valore dell'altezza e della larghezza
    return min_dimension,min_couple[0],min_couple[1]

#funzione per trovare la dimensione massima
def find_max_size(path):
    max_dimension=0
    max_couple=(0,0)

    #cerco per ogni immagine nelle sottocartelle della cartella generale la dimensione corrente e confronto con la minima
    for folder in os.listdir(path):
        #print("folder:", folder)
        for item in os.listdir(path+"/"+folder):
            #print(os.path.join(path+"/"+folder,item))
            image = cv2.imread(os.path.join(path+"/"+folder,item))

            annotated_image=image.copy()
            h,w,c=annotated_image.shape

            current_dimension=h*w
            current_couple=(h,w)

            #print("current_dimension:", current_dimension,"current_couple:", current_couple)
            if(current_dimension>max_dimension):
                max_dimension=current_dimension
                max_couple=current_couple

    #resituisco la dimensione massima, l'altezza e la larghezza
    return max_dimension,max_couple[0],max_couple[1]


#cerco in tutto il dataset la più piccola dimensione l'altezza e la larghezza ad essa associata
smallest_dimension, smallest_h, smallest_w=find_min_size(path_dataset)
print("Smallest dimension: ",smallest_dimension,", Smallest h: ", smallest_h, ", Smallest w: ", smallest_w)


#cerco in tutto il dataset la più grande dimensione l'altezza e la larghezza ad essa associata
biggest_dimension, biggest_h, biggest_w=find_max_size(path_dataset)
print("Biggest dimension: ",biggest_dimension,", Biggest h: ", biggest_h, ", Biggest w: ", biggest_w)
```
